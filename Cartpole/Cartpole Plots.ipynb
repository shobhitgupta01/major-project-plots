# Plots for CartPole Game

## Deep Q learning - Training data

```python
% matplotlib inline 

import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('results_DQN_model_1.csv')


#Storing the different columns in variables
episodes = df['episode']
epsilon = df['epsilon']
score = df['score']
average_score=df['average_score']
total_reward=df['total_reward']
average_reward=df['average_reward']


plt.title("Scores for Deep Learning Model(Training)")
plt.xlabel("Episode")
plt.ylabel("Score")
plt.plot(episodes,score,'c-')

plt.savefig('outputs/dql_score1.png')
plt.show()

```

```python
#Average Scores
plt.title("Average scores for Deep Learning Model(Training)")
plt.xlabel("Episode")
plt.ylabel("Average Score")
plt.plot(episodes,average_score,'b-')

plt.savefig('outputs/dql_avg_score.png')
plt.show()
```

```python
plt.title("Decay of Exploration Factor")
plt.xlabel("Episode")
plt.ylabel("Exploration Factor")
plt.plot(episodes,epsilon,'r--')

plt.savefig('outputs/dql_decay.png')
plt.show()
```

```python
plt.title("Total rewards for Deep Learning Model(Training)")
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.plot(episodes,total_reward,'g-')

plt.savefig('outputs/dql_rewards1.png')
plt.show()
```

```python
plt.title("Average rewards for Deep Learning Model(Training)")
plt.xlabel("Episode")
plt.ylabel("Average Reward")
plt.plot(episodes,average_reward,'b-')

plt.savefig('outputs/dql_avg_rewards1.png')
plt.show()
```

```python
#Combining the graphs for Deep Learning Training Data

fig = plt.figure(figsize=(12,8))

ax1 = plt.subplot2grid((3,2),(0,0),rowspan=1,colspan=1)
ax2 = plt.subplot2grid((3,2),(0,1),rowspan=1,colspan=1)
ax3 = plt.subplot2grid((3,2),(1,0),rowspan=1,colspan=1)
ax4 = plt.subplot2grid((3,2),(1,1),rowspan=1,colspan=1)
ax5 = plt.subplot2grid((3,2),(2,0),rowspan=1,colspan=2)


#Total Scores
ax1.set_title("Scores for Deep Learning Model(Training)")
ax1.set_xlabel("Episode")
ax1.set_ylabel("Score")
ax1.plot(episodes,score,'c-')


#Average Scores
ax2.set_title("Average scores for Deep Learning Model(Training)")
ax2.set_xlabel("Episode")
ax2.set_ylabel("Average Score")
ax2.plot(episodes,average_score,'b-')


#Exploration Factor
ax3.set_title("Decay of Exploration Factor")
ax3.set_xlabel("Episode")
ax3.set_ylabel("Exploratino Factor")
ax3.plot(episodes,epsilon,'r--')


#Total Rewards
ax4.set_title("Total rewards for Deep Learning Model(Training)")
ax4.set_xlabel("Episode")
ax4.set_ylabel("Total Reward")
ax4.plot(episodes,total_reward,'g-')


#Average rewards

ax5.set_title("Average rewards for Deep Learning Model(Training)")
ax5.set_xlabel("Episode")
ax5.set_ylabel("Average Reward")
ax5.plot(episodes,average_reward,'y-')

plt.tight_layout()
plt.show()
plt.savefig('outputs/dql_1.png')
```

## Comparison of Deep Q learning, Q learning and Random action (testing)


```python
#Importing the random action Data
df5 = pd.read_csv('test_results_random_guessing.csv')


#Storing the different columns in variables
episode5 = df5['episode']
epsilon5 = df5['epsilon']
score5 = df5['score']
average_score5=df5['average_score']
total_reward5=df5['total_reward']
average_reward5=df5['average_reward']
```

```python
#Scores

fig = plt.figure(figsize=(15,6))

ax1 = plt.subplot2grid((1,1),(0,0))

ax1.set_title("Comparison of Scores")
ax1.set_xlabel("Episode")
ax1.set_ylabel("Score")

ax1.plot(episodes2,score2,'c-',label="Deep Q-learning",linewidth=4)

ax1.plot(episode4,score4,'r-.',label="Q-learning",linewidth=4)

ax1.plot(episode5,score5,'b-',label="Random Guess",linewidth=4)

ax1.legend()

plt.tight_layout()
plt.savefig('outputs/cmp_score.png')
plt.show()
```
